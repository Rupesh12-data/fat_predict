```python
paq = paq[['SEQN','PAD680']]

def activity_level(minutes):

    if minutes < 240:
        return '0'

    elif minutes <= 480:
        return '1'

    else:
        return '2'

paq['Activity_Level'] = paq['PAD680'].apply(activity_level)

paq = paq[['SEQN','Activity_Level']]
```

```python
for df in [bmx, paq, demo]:
    print(df["SEQN"].dtype)
    print(df["SEQN"].isnull().sum())
    
    df["SEQN"] = df["SEQN"].astype("Int64")
    print(df["SEQN"].dtype)
```
